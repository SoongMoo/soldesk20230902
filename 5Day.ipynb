```python
rainfall = pd.read_csv('https://raw.githubusercontent.com/SoongMoo/soldesk20230902/main/%EC%B0%B8%EA%B3%A0%EC%9E%90%EB%A3%8C/NumPy/data/Seattle2014.csv');
rainfall
rainfall['PRCP']
inches = np.array(rainfall['PRCP'] / 254)
inches
# 0.5인치이상 온 날은 몇일입니까?
inches > 0.5
print(np.sum(inches > 0.5))
# 0.5인치이상 온 날의  총 강수량은?
print(inches[inches > 0.5].sum())
print(np.sum(inches[inches > 0.5]))

# 강수량이 0.5에서 0.8 사이인 날은 몇일이나 되나요?
#print((inches >= 0.5) & (inches <= 0.8))
print(((inches >= 0.5) & (inches <= 0.8)).sum())

# 강수량이 0.5에서 0.8 사이인 날의 총 강수량은?
#print(inches[(inches >= 0.5) & (inches <= 0.8)])
print(inches[(inches >= 0.5) & (inches <= 0.8)].sum())

#강수량이 0.5미만 0.8초과한 날은?
(~((inches >= 0.5) & (inches <= 0.8))).sum()
((inches < 0.5)|(inches> 0.8)).sum()
print(np.sum(~((inches >= 0.5) & (inches <= 0.8))))
print(np.sum(((inches < 0.5)|(inches> 0.8))))

#강수량이 0.5미만 0.8초과한 날의 강수량은?
print(inches[~((inches >= 0.5) & (inches <= 0.8))].sum())
print(inches[((inches < 0.5)|(inches> 0.8))].sum())

#비가 안온 날
(inches == 0).sum()
np.sum((inches == 0))
#비가 온 날
(inches != 0).sum()
np.sum(inches != 0)
```
